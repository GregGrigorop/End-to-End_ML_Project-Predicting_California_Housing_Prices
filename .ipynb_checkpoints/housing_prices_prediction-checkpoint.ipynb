### Creating a Test Set

__Note__: We will utilise the random number's generator seed to ensure that in case our program is run again the same test set will be generated (this way our algorithm will not get to see the whole dataset; the same shuffled indices will be produced every time).

```python
np.random.seed(42)
```

```python
# defining a function for creating a test set "manually" (for illustrative purposes, sklearn already has a function to this purpose)
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    train_indices = shuffled_indices[test_set_size:]
    test_indices = shuffled_indices[:test_set_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
# example
train_set, test_set = split_train_test(housing, 0.3)
print(len(train_set))
print(len(test_set))
```

__Note__: We will define a function to ensure our test set will remain consistent across multiple runs, even if we refresh our dataset. That means we will have a stable train/test split even after updating the dataset, and the way to do that is by using a unique and immutable identifier for each instance, one that will help us assess whether an instance should be part of the test set or not.

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32
```

```python
def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')
```

**Note**: Since we are using the row index as a unique identifier we must make sure that new data gets appended to the end of the dataset and that no row ever gets deleted. In case that is not possible we can try to use the most stable attributes to construct a unique identifier, e.g. the longitude and latitude of districts.

```python
housing_with_id["id"] = housing['longitude'] * 1000 + housing['latitude']
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'id')
```

__Note__: We will perform **_stratified sampling_** based on the income category.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

We will now create a DataFrame presenting the **_sampling bias comparison_** between _stratified_ and purely _random sampling_.

```python
# function that returns the income category proportions for a dataset
def income_cat_proportions(data):
    return data['income_cat'].value_counts() / len(data)

# generating the training/test sets again so that the 'income_cat' feature is present 
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

# creating the DataFrame
compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Random": income_cat_proportions(test_set),
    "Stratified": income_cat_proportions(strat_test_set)
}).sort_index()

compare_props["Random-%Error"] = 100 * compare_props['Random'] / compare_props['Overall'] - 100
compare_props["Strat-%Error"] = 100 * compare_props['Stratified'] / compare_props['Overall'] - 100
compare_props
```

__Note__: As we can see, the test set generated with stratified sampling has income category proportions that are much closer (almost identical) to those in the full dataset, whereas the test set produced using random sampling is skewed. This is why we will utilise the former in our project.

```python
# removing the 'income_cat' attribute to revert the data to its original state
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

### Data Cleaning

As we saw earlier the `total_bedrooms` feature has some missing values, so we need to deal with this. We have three options:

1. We can get rid of the corresponding districts.
2. We can get rid of the attribute altogether.
3. We can set the missing values to some value (e.g. zero, mean, median). That value needs to be computed on the training set.

```python
incomplete_rows = housing[housing.isnull().any(axis=1)]
print("Number of incomplete rows: ", len(incomplete_rows))
incomplete_rows.head()
```

We can take care of missing values using sklearn's `SimpleImputer`. We will replace the missing values of `total_bedrooms` with the median of this attribute. Before fitting the `imputer` instance to the training data we need to remove all the text attributes (we only have one here), as the median can only be computed on numerical features.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')

# removing the only text attribute we have
housing_num = housing.drop('ocean_proximity', axis=1)

# training the imputer
imputer.fit(housing_num)
```

__Note__: Only the `total_bedrooms` feature had missing values, but we applied the imputer to all the numerical features. This is a safer approach on general since we cannot be sure that there won't be any missing values in new data after a system we are building goes live.

```python
# checking the calculated median values
print(imputer.statistics_)

# calculating the medians "manually" to compare
print(housing_num.median().values)

# imputation strategy selected
imputer.strategy
```

We will now use our "trained" imputer to transform the training set by replacing missing values with the learned medians.

```python
X = imputer.transform(housing_num)

housing_tr = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index)
housing_tr.head()
```

```python
# the originally incomplete rows after the imputation
housing_tr.loc[incomplete_rows.index.values]
```

### Handling text and categorical features

In our dataset we have only one text attribute, `ocean_proximity`.

```python
housing_cat = housing[['ocean_proximity']]
housing_cat.head(10)
```

```python
housing_cat['ocean_proximity'].value_counts()
```

__Note__: As we can see we are not dealing with arbitrary text, we have a limited number of possible values, each of which represents a category. This indicates that this is a _categorical attribute_. Most Machine Learning algorithms prefer to work with numbers, so we will convert these categories from text to numbers by using sklearn's `OrdinalEncoder` class.

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(housing_cat)
housing_cat_encoded[:10]
```

```python
# viewing the list of categories
ordinal_encoder.categories_
```

This representation is not ideal, since ML algorithms will assume that two nearby values are more similar than two distant values, and our categories are not ordered (category 0 is more similar to category 4 and very different from category 1). To fix this issue we will apply **_one-hot encoding_**, meaning we will create one binary attribute per category: one attribute will be equal to 1 (hot), while the others will be 0 (cold). The class to use here would be the `OneHotEncoder` class, as it converts categorical values into one-hot vectors.

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_cat_1hot = cat_encoder.fit_transform(housing_cat)
# viewing the categories
print(cat_encoder.categories_)
housing_cat_1hot
```

The output is a SciPy sparse matrix (not a NumPy array) and this is very useful as it frees up a lot of memory. If we need to convert it to a (dense) NumPy array no matter what we can do that as follows:

```python
housing_cat_1hot.toarray()
```

We will now create a __custom transformer__ that adds extra, combined features.

```python
from sklearn.base import BaseEstimator, TransformerMixin

col_names = ["total_rooms", "total_bedrooms", "population", "households"]
rooms_ix, bedrooms_ix, population_ix, households_ix = [housing.columns.get_loc(c) for c in col_names]

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # add_bedrooms_per_room is the only hyperparameter of this transformer
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)

# housing_extra_attribs is a NumPy array, we've lost the column names (that's a problem
# with Scikit-Learn); to generate a DataFrame, we can run this:
housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(housing.columns) + ["rooms_per_household", "population_per_household"],
    index=housing.index)
housing_extra_attribs.head()
```

### Transformation pipelines

As we can see, there are many data transformation steps that need to be executed in the right order. We will use sklearn's `Pipeline` class to deal with such transformations sequences. We will first build a small **pipeline for preprocessing the numerical features**.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy='median')),
    ("attribs_adder", CombinedAttributesAdder()),
    ("std_scaler", StandardScaler())
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
housing_num_tr
```

We will now use a **single transformer able to handle all columns**, numerical and categorical, so as to apply the appropriate transformations to each column without handling categorical and numerical features separately.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs)
])

housing_prepared = full_pipeline.fit_transform(housing)
print(housing_prepared.shape)
housing_prepared
```

## Selecting and training a model

After getting the data, exploring it, sampling a training and a test set, and building transformation pipelines to automatically clean up and prepare our data for ML algorithms, we are now ready to select and train a Machine Learning model.

### Training and evaluating on the training set

We will first train a __Linear Regression__ model and try it out on a few instances from the training set.

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print(f"Predictions: {lin_reg.predict(some_data_prepared)}")
print(f"Labels: {list(some_labels)}") # actual values
```

We can see that our predictions are not very accurate. We will now measure this regression model's RMSE on the entire training set.

```python
from sklearn.metrics import mean_squared_error

housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
print("RMSE (Linear Regression):", lin_rmse)
```

```python
print(f"Descriptive statistics of our target attribute '{housing_labels.name}'")
housing_labels.describe()
```

__Note__: As we can see the median housing values of most districts lie (approximately) within the range of \\$120,000 and \\$264,000, so a typical prediction error of \\$68,628 is not a great outcome. Here we have an example of a model _underfitting_ the training data. We will now try a __Decision Tree Regression__ model, a more powerful model that is capable of finding complex nonlinear relationships in a dataset.

```python
from sklearn.tree import DecisionTreeRegressor

#training the model
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(housing_prepared, housing_labels)
```

```python
# evaluating on the training set
housing_predictions = tree_reg.predict(housing_prepared)
tree_rmse = mean_squared_error(housing_labels, housing_predictions, squared=False)
print("RMSE (Decision Tree):", tree_rmse)

from sklearn.metrics import mean_absolute_error

tree_mae = mean_absolute_error(housing_labels, housing_predictions)
print("MAE (Decision Tree):", tree_mae)
```

__Note__: There is no error at all, so this model is most probably badly _overfitting_ the training dataset. We will confirm that this is the case using cross-validation, so we will use a part of the training set for training and a part for model validation. That is the way to go, since we should not touch the test set for for evaluation purposes until we are ready to productionise our model.

### Cross-validation

We will utilise sklearn's _K-fold cross-validation_ feature to carry out better evaluation.

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

**Note**: Sklearn's cross-validation features expect a utility function (not a cost function), so the scoring function provides the opposite of the RMSE; this explains the negative sign in the last line of code in the previous cell.

```python
def display_scores(scores):
    print(f"Scores: {scores} \nMean: {scores.mean()} \nStandard Deviation: {scores.std()}")
```

```python
print(f"Cross-validation - Decision Tree Regression model\n")
display_scores(tree_rmse_scores)
```

The mean of the scores provides an estimate of model performance and the standard deviation of the scores is a measure of how precise this estimate is. So, judging by the output, the Decision Tree seems to be actually performing worse than the Linear Regression model. Just to be sure, we will calculate the same scores for the Linear Regression model as well.

```python
# cross-validation for linear regression
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                            scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
```

```python
print(f"Cross-validation - Linear Regression model\n")
display_scores(lin_rmse_scores)
```

```python
pd.Series(lin_rmse_scores).describe()
```

We now know for certain that the Decision Tree model overfits the training data so badly that it ends up performing worse than Linear Regression. We will try another model at this point, a __Random Forest Regression__ model.

```python
from sklearn.ensemble import RandomForestRegressor

# training the model
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(housing_prepared, housing_labels)
```

```python
# evaluating on the training set
housing_predictions = forest_reg.predict(housing_prepared)
forest_rmse = mean_squared_error(housing_labels, housing_predictions, squared=False)
print("RMSE (Random Forest):", forest_rmse)
```

```python
# cross-validation
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels,
                               scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
```

```python
print(f"Cross-validation - Random Forest Regression model\n")
display_scores(forest_rmse_scores)
```

__Note__: The Random Forest seems like the best solution so far. However, its score on the training set (approx. 18,604) is still much lower than on the validation sets, so this model is still overfitting the training set.

We will try out a couple of __Support Vector Machine__ models as well to see how they perform.

```python
from sklearn.svm import SVR

for kernel_ in ["linear", "poly"]:

    print(f"{kernel_.capitalize() if kernel_ == 'linear' else kernel_.capitalize()+'nomial'} kernel\n")
    
    # training the model
    svm_reg = SVR(kernel=kernel_)
    svm_reg.fit(housing_prepared, housing_labels)

    # evaluating on the training set
    housing_predictions = svm_reg.predict(housing_prepared)
    svm_rmse = mean_squared_error(housing_labels, housing_predictions, squared=False)
    print("RMSE (Training Set):", svm_rmse, "\n")

    # cross-validation
    print(f"Cross-validation\n")
    svm_scores = cross_val_score(svm_reg, housing_prepared, housing_labels,
                                scoring="neg_mean_squared_error", cv=10)
    svm_rmse_scores = np.sqrt(-svm_scores)
    display_scores(svm_rmse_scores)
    if kernel_ == 'linear':
        print("\n")
```

It is clearly visible that the SVM models are the worst performers of all the models we have tried so far, even when evaluating them on the training set, with the SVM model with the polynomial kernel performing worse than the one with the linear kernel.

### Fine-tuning our models

We will use Scikit-Learn's `GridSearchCV` to search for the best combination of hyperparameters for the Random Forest Regression model. We will have 90 rounds of training (since we will try 18 combinations of RandomForestRegressor hyperparameter values with five-fold cross-validation).

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # 1st stage: evaluation of 12 combinations of hyperparameter values
    {"n_estimators": [3,10,30], "max_features": [2,4,6,8]},
    # 2nd stage: evaluation 6 combinations of hyperparameters, with the 'bootstrap' hyperparameter set to False
    {"bootstrap": [False], "n_estimators": [3,10], "max_features": [2,3,4]}
]

forest_reg = RandomForestRegressor(random_state=42)
# training the model
grid_search = GridSearchCV(forest_reg, param_grid,
                           scoring="neg_mean_squared_error",
                          cv=5, return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```

```python
# best combination of parameters found
print(f"Best hyperparameters: {grid_search.best_params_}")
# best estimator found
print(f"Best estimator: {grid_search.best_estimator_}")
```

```python
pd.DataFrame(grid_search.cv_results_)
```

We will also check the __evaluation scores__ (scores of all hyperparameter combinations tested during the grid search).

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

__Note__: The best solution is obtained by setting the `max_features` and `n_estimators` hyperparameters to 8 and 30 respectively. The resulting RMSE score for this combination is approx. 49,682 and it is slightly better than the score we achieved (50,182) using the default hyperparameter values.

We will also try out the __Randomized Search__ tuning approach to check whether we can discover an even better combination of hyperparameters for the Random Forest Regression model.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
    "n_estimators": randint(low=1, high=200),
    "max_features": randint(low=1, high=8)
}

forest_reg = RandomForestRegressor(random_state=42)
# training the model
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs, n_iter=10,
                               cv=5, scoring="neg_mean_squared_error", random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
print(f"Best hyperparameters {rnd_search.best_params_}")
print(f"Best estimator: {rnd_search.best_estimator_}")
```

Using randomized search we managed to obtain an even better combination of hyperparameter values `max_features` = 7 and `n_estimators` = 180) with a RMSE of approx. 49,151.

### Evaluating on the test set

After tweaking our models we finally have a system that performs sufficiently well. We will now **evaluate the best model that grid search generated on the test set**.

```python
# selecting the best model we got using GridSearchCV
final_model_gr = grid_search.best_estimator_

# getting predictors and labels from our test set
X_test = strat_test_set.drop('median_house_value',axis=1)
y_test = strat_test_set['median_house_value'].copy()

# transforming the data (with transform(), not fit_transform() as we do not want to fit the test set)
X_test_prepared = full_pipeline.transform(X_test)

# evaluating on the test set
final_predictions_gr = final_model_gr.predict(X_test_prepared)
final_rmse_gr = mean_squared_error(y_test, final_predictions_gr, squared=False)
print("Test RMSE (Random Forest/GridSearchCV):", final_rmse_gr)
```

__Note__: Just a point estimate of the generalization error is not sufficient, so we will now get an idea of how precise that estimate is by calculating the **95% confidence interval for the generalization error** (the test RMSE).

```python
from scipy import stats

confidence = 0.95
squared_errors_gr = (y_test - final_predictions_gr) ** 2
ci_95_gr = np.sqrt(stats.t.interval(confidence, len(squared_errors_gr) -1,
                        loc=squared_errors_gr.mean(), 
                        scale=stats.sem(squared_errors_gr)))
print(f"95% confidence interval for test RMSE (Random Forest/Grid search):")
print(list(ci_95_gr))
```

```python
print(f"Difference between the Test RMSE and the lower bound of the 95% conf. interval:\n")
print(final_rmse_gr - np.sqrt(stats.t.interval(confidence, len(squared_errors_gr) -1,
                        loc=squared_errors_gr.mean(), 
                        scale=stats.sem(squared_errors_gr)))[0],"\n")
print("Difference between the upper bound of the 95% conf. interval and the Test RMSE:")
-final_rmse_gr + np.sqrt(stats.t.interval(confidence, len(squared_errors_gr) -1,
                        loc=squared_errors_gr.mean(), 
                        scale=stats.sem(squared_errors_gr)))[1]
```

```python
# computing the interval manually
m = len(squared_errors_gr)
mean = squared_errors_gr.mean()
tscore = stats.t.ppf((1 + confidence)/2, df = m -1)
tmargin = tscore * squared_errors_gr.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
```

```python
# using z-scores instead of t-scores
zscore = stats.norm.ppf((1 + confidence)/2)
zmargin = zscore * squared_errors_gr.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
```
